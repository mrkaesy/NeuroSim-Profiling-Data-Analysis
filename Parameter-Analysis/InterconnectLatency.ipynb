```python
import pandas as pd
import matplotlib.pyplot as plt

# Data extracted from the LaTeX table
data = {
    'MemCellType': [1]*5 + [2]*5,
    'numRowSubArray/numColSubArray': [64, 128, 256, 512, 1024]*2,
    'Layer1': [24095, 23752.5, 23684.7, 25506.1, 36382.1, 41004.7, 23317.4, 23219.6, 23214.7, 23603.6],
    'Layer2': [16888.7, 16897.4, 16976.4, 18421.6, 26276.7, 45996, 31381.6, 29550.9, 16766.7, 17047.5],
    'Layer3': [13491.8, 13526, 13603, 14776, 21076.6, 37670.3, 26165.8, 24579.9, 13448.6, 13673.8],
    'Layer4': [12445, 12475.7, 12546.4, 13627.8, 19438.8, 35613.6, 23863.3, 23544, 12403.5, 12611.3],
    'Layer5': [4612.62, 4625.92, 4653.08, 5055.17, 7210.75, 13164.5, 8966.44, 8693.61, 4601.04, 4678.12],
    'Layer6': [8638.63, 8665.16, 8716.87, 9471.02, 13509.6, 25174.7, 16808.6, 16807.6, 8620.19, 8764.59],
    'Layer7': [195.462, 195.636, 196.589, 213.364, 304.345, 362.699, 192.052, 192.729, 194.197, 197.45],
    'Layer8': [172.405, 172.935, 173.968, 189.019, 269.619, 323.531, 169.76, 170.552, 172.039, 174.92]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Setting a smaller figure size for the plot
plt.figure(figsize=(10, 6))

# Distinguishing between MemCellType 1 and 2 with different line styles
line_styles = ['-', '--']
memcell_legend = ['MemCellType 1', 'MemCellType 2']

# Mapping symbols to layers
symbol_map = {1: 'o', 2: '^', 3: 's', 4: 'd', 5: 'x', 6: '*', 7: '+', 8: 'h'}

# Plotting
for mem_type in df['MemCellType'].unique():
    subset = df[df['MemCellType'] == mem_type]
    for i, layer in enumerate(['Layer1', 'Layer2', 'Layer3', 'Layer4', 'Layer5', 'Layer6', 'Layer7', 'Layer8']):
        plt.plot(subset['numRowSubArray/numColSubArray'], subset[layer], line_styles[mem_type - 1] + symbol_map[i + 1], label=f'Layer {i + 1} (Type {mem_type})')

# Adding legend with MemCellType distinction
plt.legend()

# Adding plot title and labels
plt.title('Interconnect Latency Across Layers and Configurations')
plt.xlabel('numRowSubArray/numColSubArray')
plt.ylabel('Interconnect Latency (ns)')
plt.xscale('log')
plt.yscale('log')
plt.xticks(df['numRowSubArray/numColSubArray'].unique(), labels=df['numRowSubArray/numColSubArray'].unique())
plt.grid(True, which="both", ls="--")

# Note: To display the plot in your environment, uncomment the line below
# plt.show()
```
