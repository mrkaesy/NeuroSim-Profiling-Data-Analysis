```python
import ipywidgets as widgets

# Define global variables for memcell type, subarray size, and tech node size
memcell_type = 1
subarray_size = 64
tech_node_size = 10
model_name ="VGG8"


# Define dropdown widgets
memcell_type_dropdown = widgets.Dropdown(
    options=[('Type1',1), ('Type2',2)],  # Example options
    description='MemCell Type:',
)

subarray_size_dropdown = widgets.Dropdown(
    options=[64, 128, 256, 512, 1024, 2048],  # Example options
    description='Subarray Size:',
)

# Dropdown menu for Tech Node Size
tech_node_dropdown = widgets.Dropdown(
    options=[('5nm', 5),('7nm', 7),('10nm', 10), ('14nm', 14), ('22nm', 22)],  # Example options, you can customize as needed
    description='Tech Node Size:',
)
# Dropdown menu for Model
model_dropdown = widgets.Dropdown(
    options=[('VGG8', "VGG8"), ('DenseNet40', "DenseNet40")],  # Example options, you can customize as needed
    description='Select Model :',
)

# Define a function to update the global variables
def update_global_variables(change):
    global memcell_type, subarray_size, tech_node_size,model_name
    memcell_type = memcell_type_dropdown.value
    subarray_size = subarray_size_dropdown.value
    tech_node_size = tech_node_dropdown.value
    model_name =  model_dropdown.value

# Attach the function to the dropdown widgets' observe method
memcell_type_dropdown.observe(update_global_variables, names='value')
subarray_size_dropdown.observe(update_global_variables, names='value')
tech_node_dropdown.observe(update_global_variables, names='value')
model_dropdown.observe(update_global_variables,names='value')

# Display the dropdowns
display(widgets.VBox([memcell_type_dropdown, subarray_size_dropdown, tech_node_dropdown,model_dropdown]))
# Define dropdown widgets for prediction target and performance metrics
prediction_target_dropdown = widgets.Dropdown(
    options=['ReadLatency', 'ReadDyanamicEnergy', 'LeakagePower', 'LeakageEnergy', 'BufferLatency', 'BufferReadDynamicEnergy', 'ICLatency', 'ICReadDynamicEnergy'],
    description='Prediction Target:'
)

performance_metric_dropdown = widgets.Dropdown(
    options=['Energy Efficiency TOPS/W (Pipelined Process)', 'Throughput TOPS (Pipelined Process)', 'Throughput FPS (Pipelined Process)', 'Compute efficiency TOPS/mm^2 (Pipelined Process)'],
    description='Performance Metric:'
)

# Define global variables to store the selected values
selected_prediction_target = "ReadLatency"
selected_performance_metric = "Energy Efficiency TOPS/W (Pipelined Process)"

# Define a function to update the global variables with the selected values
def update_selected_values(change):
    global selected_prediction_target, selected_performance_metric
    selected_prediction_target = prediction_target_dropdown.value
    selected_performance_metric = performance_metric_dropdown.value

# Attach the function to the dropdown widgets' observe method
prediction_target_dropdown.observe(update_selected_values, names='value')
performance_metric_dropdown.observe(update_selected_values, names='value')

# Display the dropdown menus
display(widgets.VBox([prediction_target_dropdown, performance_metric_dropdown]))
```
